### Clients that are in the data events table

```python
unique_clients_events = data_events["USER_CLIENT_NUMBER"].unique()
len(unique_clients_events)
```

### Clients that have made purchases

```python
unique_clients_purchases = purchase_events["USER_CLIENT_NUMBER"].unique()
len(unique_clients_purchases)
```

### Clients that have data events but have not made purchases

```python
clients_without_purchases = [client_id for client_id in unique_clients_events if client_id not in unique_clients_purchases]
len(clients_without_purchases)
```

### Clients that made purchases but do not have data events
These clients are interesting, why don't they have any data available in the data events? Is that because we received incomplete data or how is this else possible?

```python
clients_with_purchases_without_data_events =[client_id for client_id in unique_clients_purchases if client_id not in unique_clients_events] 
clients_with_purchases_without_data_events_table = purchase_events[purchase_events["USER_CLIENT_NUMBER"].isin(clients_with_purchases_without_data_events)]
```

```python
import numpy as np
total_amount_clients_without_data_events = np.sum(clients_with_purchases_without_data_events_table["AMOUNT"])
print(f"{len(clients_with_purchases_without_data_events)} clients have spend A total of {np.round(total_amount_clients_without_data_events/1000000,decimals=1)} million euro and these clients do not have any data events available")
```

# Clients that we use for analysis. Let us explore the clients that have made purchases and also have recorded data events 

### Clients that have made purchases and have data events
These are the clients that we can properly analyze because about these clients we can analyze their behavior on the website

```python
clients_with_purchases_and_data_events = [client_id for client_id in unique_clients_purchases if client_id in unique_clients_events]
clients_with_purchases_and_data_events_table = purchase_events_processed[purchase_events_processed["USER_CLIENT_NUMBER"].isin(clients_with_purchases_and_data_events)]
```

```python
total_amount_clients_with_data_events = np.sum(clients_with_purchases_and_data_events_table["AMOUNT"])
print(f"{len(clients_with_purchases_and_data_events)} cients have spend a total of {np.round(total_amount_clients_with_data_events/1000000,decimals=1)} million euro who also have data events available")
```
