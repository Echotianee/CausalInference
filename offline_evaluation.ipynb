```python
import os
def save_client_data_to_folder(data_events_table, purchase_events_table, client_id):
    folder_main = str(client_id)
    purchase_file = "purchases.csv"
    data_file = "events.csv"
    
    client_data_events = data_events_table[data_events_table["USER_CLIENT_NUMBER"]==client_id]
    client_purchases = purchase_events_table[purchase_events_table["USER_CLIENT_NUMBER"]==client_id]
    
    client_folder = os.path.join("client_data", folder_main)
    
    if not os.path.isdir(client_folder):
        os.mkdir(client_folder)
        
    client_data_events.to_csv(os.path.join(client_folder, data_file), sep="|")
    client_purchases.to_csv(os.path.join(client_folder, purchase_file), sep ="|")
    
save_client_data_to_folder(data_events, purchase_events, 471557)
```

```python
unique_clients_purchases = purchase_events["USER_CLIENT_NUMBER"].unique()
unique_clients_events = data_events["USER_CLIENT_NUMBER"].unique()

clients_with_purchases_and_data_events = [client_id for client_id in unique_clients_purchases if client_id in unique_clients_events]
```

```python
def data_events_to_categories_dict_creation(data_events):
    transformation_dict = {}
    for col in ["PAGE_NAME", "PAGE_SECTION", "PRODUCT_TYPE", "DEVICE_INFO_BRAND", "DEVICE_INFO_TYPE", "DEVICE_INFO_BROWSER", "USER_SALES_GROUP", "USER_SEGMENT", "USER_SALES_DISTRICT", "EVENT"]:
        print(data_events.keys())
        unique_col_values = data_events[col].unique()
        
        category_number = 0
        transformation_dict[col] = {}
        for col_value in unique_col_values:
            transformation_dict[col][col_value] = category_number
            category_number +=1
    return transformation_dict

transformation_dict = data_events_to_categories_dict_creation(data_events)
transformation_dict
```

```python
for key, dict in transformation_dict.items():
    print(key)
    print(len(dict))
```

```python
def get_inverse_transformation_dict(transformation_dict):
    #inverse of transformation dict
    inverse_transformation_dict = {}
    for col_name, dict_vals in transformation_dict.items():
        inverse_transformation_dict[col_name] = {}
        for dict_val, cat_num in dict_vals.items():
            inverse_transformation_dict[col_name][cat_num] = dict_val
            
    return inverse_transformation_dict

```

```python
def transform_df_to_categories(events_table, to_categories=True):
    transformation_dict = data_events_to_categories_dict_creation(events_table)
    inverse_transformation_dict = get_inverse_transformation_dict(transformation_dict)
    if to_categories:
        for column, mapping in transformation_dict.items():
            events_table[column] = events_table[column].map(lambda x: mapping[x])
    else:
        for column, mapping in inverse_transformation_dict.items():
            events_table[column] = events_table[column].map(lambda x: mapping[x])
    return events_table
        
# Replace categorical values using map

    
transform_df_to_categories(data_events, to_categories=False)
```
