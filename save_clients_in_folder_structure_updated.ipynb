```python
def save_client_data_to_folder(data_events_table, purchase_events_table, client_id):
    folder_main = str(client_id)
    purchase_file = "purchases.csv"
    data_file = "events.csv"
    
    client_data_events = data_events_table[data_events_table["USER_CLIENT_NUMBER"]==client_id]
    client_purchases = purchase_events_table[purchase_events_table["USER_CLIENT_NUMBER"]==client_id]
    
    client_folder = os.path.join("client_data", folder_main)
    
    if not os.path.isdir(client_folder):
        os.mkdir(client_folder)
        
    client_data_events.to_csv(os.path.join(client_folder, data_file), sep="|", index=False)
    client_purchases.to_csv(os.path.join(client_folder, purchase_file), sep ="|", index=False)
    
save_client_data_to_folder(data_events, purchase_events, 471557)
```

```python
unique_clients_purchases = purchase_events["USER_CLIENT_NUMBER"].unique()
unique_clients_events = data_events["USER_CLIENT_NUMBER"].unique()

clients_with_purchases_and_data_events = [client_id for client_id in unique_clients_purchases if client_id in unique_clients_events]
print(len(clients_with_purchases_and_data_events))
```

```python
for client in clients_with_purchases_and_data_events:
    save_client_data_to_folder(data_events, purchase_events, client)
```
