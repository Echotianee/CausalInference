```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data_events
data_events = pd.read_csv('processed_data/processed_events.csv', sep="|")

```

```python
#import weather data
data_weather = pd.read_csv('data/weather_utrecht_data_daily.csv', sep=",")
data_weather
```

```python
def select_transform_columns(df):
    """
    This function transforms the columns of the weather data to the right format
    


    Args:
        df: 

    Returns:

    """
    new_df = df.copy()
    new_df["date"] = pd.to_datetime(new_df["datetime"])
    new_df["temperature"] = new_df["feelslike"]
    
    return new_df[["date", "temperature","precipcover","precip"]]

data_weather_used_columns = select_transform_columns(data_weather)
```

```python
data_weather_used_columns
```

```python
for lead in range(1, 5):  # 1 to 4 days lead
    data_weather_used_columns[f'temperature_lead_{lead}'] = data_weather_used_columns['temperature'].shift(-lead)
    data_weather_used_columns[f'precipitation_coverage_lead_{lead}'] = data_weather_used_columns['precipcover'].shift(-lead)
    data_weather_used_columns[f'precipitation_amount_lead_{lead}'] = data_weather_used_columns['precip'].shift(-lead)
data_weather_used_columns
```

```python
def merge_data_events_weather(data_events, data_weather):
    """
    This function merges the data_events and data_weather dataframes on the date column
    """
    data_events["DATE"] = pd.to_datetime(data_events["DATE"])
    data_weather["date"] = pd.to_datetime(data_weather["date"])
    data_events_weather = pd.merge(data_events, data_weather, left_on="DATE", right_on="date", how="left")
    return data_events_weather

data_events_weather = merge_data_events_weather(data_events, data_weather_used_columns)
data_events_weather
```

```python
data_events_weather.to_csv('processed_data/processed_data_events_weather.csv', sep="|", index=False)
```
